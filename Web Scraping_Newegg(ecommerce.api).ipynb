put everything into for loop

```python
product_name = []
brand = []
rating = []
rating_count = []
unit_price = []
final_price = []

for result in result_items:
    
    #name
    try:
        product_name.append(result['Description']['ProductName'])
    except:
        product_name.append('n/a')
        
    #brand
    try:
        brand.append(result['ItemManufactory']['Manufactory'])
    except:
        brand.append('n/a')
        
    #rating
    try:
        rating.append(result['Review']['Rating'])
    except:
        rating.append('n/a')
        
    #rating_count
    try:
        rating_count.append(result['Review']['HumanRating'])
    except:
        rating_count.append('n/a')
    
    #unit_price
    try:
        unit_price.append(result['UnitCost'])
    except:
        unit_price.append('n/a')
        
    #final_price
    try:
        final_price.append(result['FinalPrice'])
    except:
        final_price.append('n/a')
    
```

```python
df_newegg = pd.DataFrame({'Product_Name':product_name,'Brand':brand,'Rating':rating,'Rating_count':rating_count,
                          'Unit_price':unit_price,'Final_price':final_price})
```
