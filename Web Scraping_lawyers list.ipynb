```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
#Get Request
```

```python
url ='https://www.justia.com/lawyers/california/san-francisco'
response = requests.get(url)
response
```

```python
#Soup Object
```

```python
soup = BeautifulSoup(response.content,'html.parser')
soup
```

```python
#results
results = soup.find_all('div',{'data-vars-action':'OrganicListing'})
len(results)
```

```python
#Target Necessary Data
#name
#short Bio
#specilization
#university
#address
#phone
```

```python
#put everything together inside a for loop
```

```python
name = []
short_bio = []
specilization = []
university = []
address = []
phone = []

for result in results:
    #name
    try:
        name.append(result.find('strong',{'class':'lawyer-name'}).text.strip())
    except:
        name.append('n/a')
        
    #short bio
    try:
        short_bio.append(result.find('span',{'class':'-hide-landscape-phone'}).text)
    except:
        short_bio.append('n/a')
    
    #specilization
    try:
        specilization.append(result.find('span',{'class':'-practices'}).text)
    except:
        specilization.append('n/a')
    
    #university
    try:
        university.append(result.find('span',{'class':'-hide-tablet -law-schools iconed-line-small'}).text)
    except:
        university.append('n/a')
    
    #address
    try:
        address.append(result.find('span',{'class':'-address -hide-landscape-tablet'}).text.strip().replace('\t','').replace('\n',''))
    except:
        address.append('n/a')
    
    #phone
    try:
        phone.append(result.find('strong',{'class':'-phone'}).text.replace('\n',''))
    except:
        phone.append('n/a')
    
```

```python
#create Data frame
df_lawyers = pd.DataFrame({'Lawyer_name':name,'Short_Bio':short_bio,'Specilization':specilization,
                           'University':university,'Address':address,'Phone':phone})
```

```python
#to excel
df_lawyers.to_excel('lawyers_list.xlsx',index = False)
```

# pagination -Scrape 20 pages


```python
name = []
short_bio = []
specilization = []
university = []
address = []
phone = []

for i in range(1,21):
    website = 'https://www.justia.com/lawyers/california/san-francisco?page=' + str(i)
    response = requests.get(website)
    soup = BeautifulSoup(response.content,'html.parser')
    results = soup.find_all('div',{'data-vars-action':'OrganicListing'})
    
    for result in results:
        #name
        try:
            name.append(result.find('strong',{'class':'lawyer-name'}).text.strip())
        except:
            name.append('n/a')

        #short bio
        try:
            short_bio.append(result.find('span',{'class':'-hide-landscape-phone'}).text)
        except:
            short_bio.append('n/a')

        #specilization
        try:
            specilization.append(result.find('span',{'class':'-practices'}).text)
        except:
            specilization.append('n/a')

        #university
        try:
            university.append(result.find('span',{'class':'-hide-tablet -law-schools iconed-line-small'}).text)
        except:
            university.append('n/a')

        #address
        try:
            address.append(result.find('span',{'class':'-address -hide-landscape-tablet'}).text.strip().replace('\t','').replace('\n',''))
        except:
            address.append('n/a')

        #phone
        try:
            phone.append(result.find('strong',{'class':'-phone'}).text.replace('\n',''))
        except:
            phone.append('n/a')
            
            
            
#create Data frame
df_lawyers_list= pd.DataFrame({'Lawyer_name':name,'Short_Bio':short_bio,'Specilization':specilization,
                           'University':university,'Address':address,'Phone':phone})

    
```
