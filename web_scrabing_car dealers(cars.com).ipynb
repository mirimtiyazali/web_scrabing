# import libraries

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

#HTTP requests

```python
website = requests.get('https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D')
website
```

```python
soup = BeautifulSoup(website.content,'html.parser')
soup
```

```python
results = soup.find_all('div',{'class':'vehicle-card'})
len(results)
```

put everything together inside a for loop

```python
name = []
mileage = []
dealer = []
rating = []
review_count =[]
price = []

for result in results:
    
    #name
    try:
        name.append(result.find('h2').text)
        
    except:
        name.append('n/a')  
    #milage
    try:
        mileage.append(result.find('div',{'class':'mileage'}).text)
        
    except:
        mileage.append('n/a')   
    #dealer
    try:
        dealer.append(result.find('div',{'class':'dealer-name'}).get_text().strip())
        
    except:
        dealer.append('n/a')  
    #rating
    try:
        rating.append(result.find('span',{'class':'sds-rating__count'}).text)
        
    except:
        rating.append('n/a')
    #review_count
    try:
        review_count.append(result.find('span',{'class':'sds-rating__link sds-button-link'}).text)
        
    except:
        review_count.append('n/a')
    #price
    try:
        price.append(result.find('span',{'class':'primary-price'}).text)
        
    except:
        price.append('n/a')
    
            
```

# Create Pandas Dataframe

```python
car_dealer = pd.DataFrame({'Car_Name':name,'Milage':mileage,'Dealer_name':dealer,
                           'Rating':rating,'Review_Count':review_count,'Price':price})
```

```python
car_dealer
```

# Data cleaning

```python
car_dealer['Review_Count'] = car_dealer['Review_Count'].apply(lambda x:x.strip('reviews)').strip('('))
```

```python
car_dealer.head()
```

# Ouptput in Excel

```python
car_dealer.to_excel('top21 cars list.xlsx',index=False)
```

#Data from multiple pages

```python
name = []
mileage = []
dealer = []
rating = []
review_count =[]
price = []

for i in range(1,11):
    
    #website in variable
    website = 'https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&makes[]=&stock_type=cpo&zip=60606'
    
    #request to website
    request = requests.get(website)
    
    #soup object
    soup = BeautifulSoup(request.content,'html.parser')
    
    #results
    results = soup.find_all('div',{'class':'vehicle-card'})
    
    for result in results:
    
        #name
        try:
            name.append(result.find('h2').text)

        except:
            name.append('n/a')  
        #milage
        try:
            mileage.append(result.find('div',{'class':'mileage'}).text)

        except:
            mileage.append('n/a')   
        #dealer
        try:
            dealer.append(result.find('div',{'class':'dealer-name'}).get_text().strip())

        except:
            dealer.append('n/a')  
        #rating
        try:
            rating.append(result.find('span',{'class':'sds-rating__count'}).text)

        except:
            rating.append('n/a')
        #review_count
        try:
            review_count.append(result.find('span',{'class':'sds-rating__link sds-button-link'}).text)

        except:
            review_count.append('n/a')
        #price
        try:
            price.append(result.find('span',{'class':'primary-price'}).text)

        except:
            price.append('n/a')
    
    
    
```

```python
car_dealer = pd.DataFrame({'Car_Name':name,'Milage':mileage,'Dealer_name':dealer,
                           'Rating':rating,'Review_Count':review_count,'Price':price})
```

```python
car_dealer
```

```python
car_dealer['Review_Count'] = car_dealer['Review_Count'].apply(lambda x:x.strip('reviews)').strip('('))
```
