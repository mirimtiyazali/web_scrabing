```python
#list of hotels
name = []
location = []
rating_score = []
rating = []
price = []

for result in result_item:
    #name
    try:
        name.append(result['name'])
    except:
        name.append('n/a')
        
    #location
    try:
        location.append(result['address'])
    except:
        location.append('n/a')
        
    #rating score
    try:
        rating_score.append(result['ratings']['value'])
    except:
        rating_score.append('n/a')
        
    #rating
    try:
        rating.append(result['ratings']['count'])
    except:
        rating.append('n/a')
        
    #price
    try:
        price.append(result['mrcData'][0]['finalPrice'])
    except:
        price.append('n/a')
    

```

```python
oyo_rooms=pd.DataFrame({'Hotels':name,'Location':location,'Rating_score':rating_score,'Rating':rating,'Price':price})
```
