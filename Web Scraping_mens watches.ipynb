```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
web = requests.get('https://kredoworld.in/collections/mens-watches')
web
```

```python
soup = BeautifulSoup(web.text,'html.parser')
soup
```

```python
list = soup.find_all('div',{'class':'grid-product__content'})
len(list)
```

```python
watch = soup.find('div',{'class':'grid-product__title grid-product__title--body'}).text
watch
```

```python
size = soup.find('div',{'class':'hc-size-from-tag'}).text
size
```

```python
original_price = soup.find('span',{'class':'grid-product__price--original'}).text
original_price
```

```python
after_discount = soup.find('span',{'class':'grid-product__price--savings'}).text.strip()
after_discount
```

```python
name = []
size = []
price = []
discount = []

for result in list:
    
    #watch
    try:
        name.append(result.find('div',{'class':'grid-product__title grid-product__title--body'}).text)    
    except:
        name.append('n/a')  
        
    #size
    try:
        size.append(result.find('div',{'class':'hc-size-from-tag'}).text)   
    except:
        size.append('n/a')
        
    #price 
    try:
        price.append(result.find('span',{'class':'grid-product__price--original'}).text)      
    except:
        price.append('n/a') 
        
    #discount
    try:
        discount.append(result.find('span',{'class':'grid-product__price--savings'}).text.strip())  
    except:
        discount.append('n/a')
    
```

```python
latest_watches = pd.DataFrame({'watch_name':name,'Size':size,'Price':price,'Discount':discount})
latest_watches
```

```python
latest_watches.to_excel('new_watch_collection.xlsx',index=False)
```
