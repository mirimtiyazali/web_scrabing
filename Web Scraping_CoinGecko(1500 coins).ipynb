data in single table using for loop

```python
name = []
price = []
change_1hr = []
change_24hr= []
change_7d = []
volumne_24 = []
market_cap = []

for result in coin_data:
    #name
    try:
        name.append(result.find('a',{'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).text.strip())
    except:
        name.append('n/a')
        
    #price
    try:
        price.append(result.find('td',{'class':'td-price price text-right pl-0'}).text.strip())
    except:
        price.append('n/a')
        
    #1hr change
    try:
        change_1hr.append(result.find('td',{'class':'td-change1h change1h stat-percent text-right col-market'}).text.strip())
    except:
        change_1hr.append('n/a')
        
    #24hr change
    try:
        change_24hr.append(result.find('td',{'class':'td-change24h change24h stat-percent text-right col-market'}).text.strip())
    except:
        change_24hr.append('n/a')
        
    #7d change
    try:
        change_7d.append(result.find('td',{'class':'td-change7d change7d stat-percent text-right col-market'}).text.strip())
    except:
        change_7d.append('n/a')
        
    #24 volume
    try:
        volumne_24.append(result.find('td',{'class':'td-liquidity_score lit text-right col-market'}).text.strip())
    except:
        volumne_24.append('n/a')
        
    #market cap
    try:
        market_cap.append(result.find('td',{'class':'td-market_cap cap col-market cap-price text-right'}).text.strip())
    except:
        market_cap.append('n/a')
    
    
    


```

Create Data Frame

```python
crypto_df = pd.DataFrame({'Coin':name,'Price':price,'Change_1hr':change_1hr,'Change_24hr':change_24hr,'Change_7d':change_7d,'Volume_24':volumne_24,'Market_Cap':market_cap})
```

# Pagination -get 1500 results

```python
name = []
price = []
change_1hr = []
change_24hr= []
change_7d = []
volumne_24 = []
market_cap = []


for i in range(1,16):
    #website
    website = 'https://www.coingecko.com/?page=' + str(i)
    #request
    response = requests.get(website)
    #soup object
    soup = BeautifulSoup(response.content,'html.parser')
    #information
    coin_data = soup.find('table',{'class':'sort table mb-0 text-sm text-lg-normal table-scrollable'}).find('tbody').find_all('tr')
    
    for result in coin_data:
        #name
        try:
            name.append(result.find('a',{'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).text.strip())
        except:
            name.append('n/a')

        #price
        try:
            price.append(result.find('td',{'class':'td-price price text-right pl-0'}).text.strip())
        except:
            price.append('n/a')

        #1hr change
        try:
            change_1hr.append(result.find('td',{'class':'td-change1h change1h stat-percent text-right col-market'}).text.strip())
        except:
            change_1hr.append('n/a')

        #24hr change
        try:
            change_24hr.append(result.find('td',{'class':'td-change24h change24h stat-percent text-right col-market'}).text.strip())
        except:
            change_24hr.append('n/a')

        #7d change
        try:
            change_7d.append(result.find('td',{'class':'td-change7d change7d stat-percent text-right col-market'}).text.strip())
        except:
            change_7d.append('n/a')

        #24 volume
        try:
            volumne_24.append(result.find('td',{'class':'td-liquidity_score lit text-right col-market'}).text.strip())
        except:
            volumne_24.append('n/a')

        #market cap
        try:
            market_cap.append(result.find('td',{'class':'td-market_cap cap col-market cap-price text-right'}).text.strip())
        except:
            market_cap.append('n/a')
    

    
```

```python
crypto_df = pd.DataFrame({'Coin':name,'Price':price,'Change_1hr':change_1hr,'Change_24hr':change_24hr,'Change_7d':change_7d,'Volume_24':volumne_24,'Market_Cap':market_cap})
```
